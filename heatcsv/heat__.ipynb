# ✅ Outliers Checks
Outlier checks mean looking for values in your data that are very different from the rest. These strange or unusual values are called outliers.

# 🧊 What is an outlier?
An outlier is a number that is much higher or lower than most other numbers in a dataset.

Example:

If you have these ages:
25, 26, 28, 27, 102
→ The value 102 is an outlier (most people are in their 20s)
# 📌 Why check for outliers?
They can be mistakes (e.g., someone typed "1000" instead of "100").

They can change the results of your analysis or averages.

Sometimes they are real, but still need special attention.

# 🛠️ How to find outliers in heart.csv


1. Visual Checks
Box plot: Shows outliers as dots outside the "box"

Scatter plot: Shows data points; outliers stand far from the rest

2. Math Methods
Z-Score Method:

Find how far a value is from the average (mean)

If the Z-score is above 3 or below -3, it might be an outlier

IQR Method (Interquartile Range):

Step 1: Find Q1 (25%) and Q3 (75%) of the data

Step 2: IQR = Q3 - Q1

Step 3: Values below Q1 - 1.5×IQR or above Q3 + 1.5×IQR are outliers

🔍 Example Using IQR:
Data: 10, 12, 14, 15, 16, 17, 18, 100

Q1 = 13, Q3 = 17 → IQR = 4

Outlier if:

Less than 13 - 6 = 7

More than 17 + 6 = 23
→ So, 100 is an outlier
